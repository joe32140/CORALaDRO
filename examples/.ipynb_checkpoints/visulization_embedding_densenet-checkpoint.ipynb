```python
from wilds.datasets.civilcomments_dataset import CivilCommentsDataset
from wilds.common.data_loaders import get_train_loader
from wilds.common.grouper import CombinatorialGrouper
from utils import load
import torch
```

```python
dataset = CivilCommentsDataset(root_dir="/data/wilds/data")
```

```python
val_data = dataset.get_subset('val')
```

```python
val_loader = get_train_loader('standard', val_data, batch_size=16)
```

```python
grouper = CombinatorialGrouper(dataset, ['black', 'y'])
```

```python
ckpt = torch.load('logs/best_model.pth')
```

```python
from algorithms.groupDRO import GroupDRO
```

```python
from transformers import AutoTokenizer, BertForSequenceClassification
```

```python
model = BertForSequenceClassification.from_pretrained("bert-base-uncased")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
model.load_state_dict(ckpt['algorithm'], strict=False)
```

```python
model.to('cuda:1')
model.eval()
group_list = []
hidden_list = []
count = 0
with torch.no_grad():
    for b in val_loader:
        inputs, y, group = b
        embeddings = tokenizer(list(inputs), truncation=True, padding=True, return_tensors="pt")
        embeddings.to('cuda:1')
        outputs = model(**embeddings, output_hidden_states=True, return_dict=True)
        groups = grouper.metadata_to_group(group)
        out = outputs.hidden_states[-1][:,0,:] # get CLS
        group_list.extend(groups)
        hidden_list.extend(out.cpu().numpy())
        count += 1
        if count == 100:
            break
```

```python
from yellowbrick.text import TSNEVisualizer
import numpy as np
X = np.array(hidden_list)
y = np.array(group_list)

# Create the visualizer and draw the vectors
tsne = TSNEVisualizer()
tsne.fit(X, y)
tsne.show()
```

```python
X = np.array(hidden_list)
y = np.array(group_list)

black = (y==0) | (y==1)
X_b = X[black]
y_b = y[black]
# Create the visualizer and draw the vectors
tsne = TSNEVisualizer()
tsne.fit(X_b, y_b)
tsne.show()
```
